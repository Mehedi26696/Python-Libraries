# Merge Using a DataFrame column

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando",],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

***If some data is not common***

***It's called inner join(intersection)*** 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","batimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san francisco",],
    "humidity": [65,68,72],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

***Outer join(union)***

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

***Left Join***

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

***Right Join***

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

***Merge data source***

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

***Suffixes***

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=('-left','_right'))
df3
```
