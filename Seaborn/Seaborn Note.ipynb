```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None)
```

***Change bar colors using COLOR attribute***

```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None,color='red')
```

***Change color using PALLETE attribute***

```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None,palette='dark:red')
```

***Using SATURATION parameter***

```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None,palette='dark:red',saturation=0.5)
```

***Change default aggregation method using ESTIMATOR parameter***

```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None,palette='dark:red',estimator=np.median)
```

```python
plt.figure(figsize=(15,5))
ordered_columns=mart['Product line'].sort_values().unique()
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=ordered_columns,errorbar=None,palette='dark:red',estimator=sum)
```

# Box plot

***Create a basic BOX plot on one NUMERIC variable***

```python
sns.boxplot(x='Total',data=mart,width=0.5)
```

```python
sns.boxplot(y='Total',data=mart,width=0.2)
```

***Create a basic BOX plot on one numeric variable by a CATEGORICAL variable***

```python
sns.boxplot(y='Total',x='Payment',data=mart)
```

***Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute***

```python
sns.boxplot(y='Total',x='Payment',hue='Gender',data=mart)
```

***Add MEAN marker in the box plot using Showmeans attribute and change its style using meanprops***

```python
sns.boxplot(y='Total',x='Payment',hue='Gender',data=mart,showmeans=True)
```

```python
sns.boxplot(y='Total',x='Payment',hue='Gender',data=mart,showmeans=True,meanprops={"marker": "o",
                                                                                  "markerfacecolor":"white"
                                                                                  ,"markersize":"5"
                                                                                  ,"markeredgecolor":"black"})
```

***Make HORIZONTAL box plot***

```python
sns.boxplot(x='Total',y='Payment',hue='Gender',data=mart,showmeans=True,meanprops={"marker": "o",
                                                                                  "markerfacecolor":"white"
                                                                                  ,"markersize":"5"
                                                                                  ,"markeredgecolor":"black"})
```

***Change PALETTE, LINE WIDTH etc..***

```python
sns.boxplot(y='Total',x='Payment',hue='Gender',data=mart,showmeans=True,meanprops={"marker":"o"
                                                                                  ,"markerfacecolor":"white"
                                                                                  ,"markersize":"5"
                                                                                  ,"markeredgecolor":"black"},palette='CMRmap',linewidth=5)
```

***Create box plot for EACH OF THE NUMERIC VARIABLE in the dataframe***

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=mart)
```

# Violin Plot

***Create a basic Violin Plot***

```python
sns.violinplot(y='Total',data=mart)
```

***Create a VIOLINPLOT on two categorical and one numeric variable and use split***

```python
sns.violinplot(y='Total',x='Payment',data=mart)
```

```python
sns.violinplot(y='Total',x='Payment',hue='Gender',data=mart)
```

```python
sns.violinplot(y='Total',x='Payment',hue='Gender',data=mart,split=True)
```

***Change the box in the Violin Plot to horzontal lines***

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile')
```

***Draw line for each observations in a Violin Plot***

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='stick')
```

***Change the amount of smoothing using bw attribute***

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile',bw_method=.2)
```

***Cut out the extreme values***

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile',bw_method=.2,cut=0)
```

***Expand markers in Strip plot using jitter***

```python
sns.stripplot(data=mart,y='Total',jitter=1)
```

***Draw line around the points using Linewidth***

```python
sns.stripplot(data=mart,y='Total',x='Payment',hue='Gender')
```

# Swarm Plot

***Create a basic Swarm Plot***

```python
sns.swarmplot(y='Total',data=mart)
```

```python
sns.swarmplot(x='Total',data=mart)
```

***Create Swarm Plot group by categories****

```python
sns.swarmplot(y='Total',x='Payment',data=mart)
```

```python
sns.swarmplot(y='Total',x='Payment',hue='Gender',data=mart)
```

***Showing Hues Separatly on the categorical axis***

```python
sns.swarmplot(y='Total',x='Payment',hue='Gender',data=mart,dodge=True)
```

***Styling a SWARM - Change the marker, size, color, edge color***

```python
sns.swarmplot(data=mart,y='Total',x='Payment',size=5)
```

```python
sns.swarmplot(data=mart,y='Total',x='Payment',size=5,color='red')
```

```python
sns.swarmplot(data=mart,y='Total',x='Payment',size=5,color='red',edgecolor='black')
```

```python
sns.swarmplot(data=mart,y='Total',x='Payment',size=5,color='red',edgecolor='black',marker='*')
```

***Overlay a SWARM PLOT on a BOX PLOT***

```python
sns.boxplot(data=mart,y='Total',x='Payment')
sns.swarmplot(data=mart,y='Total',x='Payment',size=5,color='black')
```

***Overlay a SWARM PLOT on a VIOLIN PLOT***

```python
sns.violinplot(data=mart,y='Total',x='Payment',inner=None,width=1)
sns.swarmplot(data=mart,y='Total',x='Payment',size=4,color='white')
```
